```python
# ---------------------------------------
# Gerekli Kütüphaneler
# ---------------------------------------
import os
import glob
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import nibabel as nib
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torch.nn import CrossEntropyLoss
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, StratifiedKFold
from collections import Counter
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import torchvision.models.video as video_models
from torch.amp import autocast, GradScaler
import albumentations as A
from albumentations.pytorch import ToTensorV2

# ---------------------------------------
# CUDA ve CUDNN Ayarları
# ---------------------------------------
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
torch.backends.cudnn.benchmark = True
torch.backends.cudnn.deterministic = False
print(f"Using device: {DEVICE}")

# ---------------------------------------
# 1) Dataset Dizini ve Sınıf Etiketleri
# ---------------------------------------
DATA_DIR = '/kaggle/input/adniunpreprocessed'  # Ham NIfTI klasörünüz
classes = {'CN': 0, 'MCI': 1, 'AD': 2}

# ---------------------------------------
# 2) Dosya Toplama ve Filtreleme
# ---------------------------------------
pattern = os.path.join(DATA_DIR, '**', '*.nii*')
all_files = glob.glob(pattern, recursive=True)
valid_paths, valid_labels = [], []

for fp in all_files:
    parent = os.path.basename(os.path.dirname(fp))
    if parent not in classes:
        continue
    if os.path.getsize(fp) == 0:
        continue
    valid_paths.append(fp)
    valid_labels.append(classes[parent])

# İkinci filtre: sadece açılabilen NIfTI
clean_paths, clean_labels = [], []
for path, label in zip(valid_paths, valid_labels):
    try:
        _ = nib.load(path)
        clean_paths.append(path)
        clean_labels.append(label)
    except:
        continue

valid_paths  = clean_paths
valid_labels = clean_labels
print(f"Total sample sayısı: {len(valid_paths)}")
print("Sınıf dağılımı:", {c: valid_labels.count(classes[c]) for c in classes})

# ---------------------------------------
# 3) Sabit Test Set Ayırma (Stratified)
# ---------------------------------------
# Önce tüm verinin %10'unu test için ayırıyoruz; kalan %90'ı k-fold için kullanacağız.
paths_trainval, paths_test, labels_trainval, labels_test = train_test_split(
    valid_paths, valid_labels, test_size=0.10, stratify=valid_labels, random_state=42
)
print(f"Train+Val / Test sayıları: {len(paths_trainval)}/{len(paths_test)}")

# ---------------------------------------
# 4) Stratified K-Fold Hazırlığı
# ---------------------------------------
n_splits = 3  # 5'ten 3'e düşürdük
skf = StratifiedKFold(n_splits=n_splits, shuffle=True)  # random_state kaldırıldı

# ---------------------------------------
# 5) Gelişmiş 3D MRI Dataset
# ---------------------------------------
class ADNI_ResNet3D_Dataset(Dataset):
    def __init__(self, paths, labels, transform=None, 
                 target_depth=64, target_height=112, target_width=112,
                 is_training=True):
        self.paths = paths
        self.labels = labels
        self.transform = transform
        self.d = target_depth
        self.h = target_height
        self.w = target_width
        self.is_training = is_training

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        # 1) Load NIfTI
        path = self.paths[idx]
        img = nib.load(path).get_fdata().astype(np.float32)

        # 2) Robust normalization (%1-%99 persentil arasını [0,1])
        p1, p99 = np.percentile(img, (1, 99))
        img = np.clip(img, p1, p99)
        img = (img - p1) / (p99 - p1 + 1e-6)

        # 3) Center-crop veya pad 3D hacim → target (D x H x W)
        D, H, W = img.shape
        
        # Depth (z ekseni) orta bölgeden target_depth al
        cd = D // 2
        hd = self.d // 2
        start_d = max(cd - hd, 0)
        end_d = start_d + self.d
        if end_d > D:
            end_d = D
            start_d = D - self.d
        patch = img[start_d:end_d, :, :]

        # Yükseklik ve genişlik için merkezden al veya pad
        ch = H // 2
        hh = self.h // 2
        start_h = max(ch - hh, 0)
        end_h = start_h + self.h
        if end_h > H:
            end_h = H
            start_h = H - self.h
        patch = patch[:, start_h:end_h, :]

        cw = W // 2
        hw = self.w // 2
        start_w = max(cw - hw, 0)
        end_w = start_w + self.w
        if end_w > W:
            end_w = W
            start_w = W - self.w
        patch = patch[:, :, start_w:end_w]

        # Eğer patch boyutları eksikse pad et
        pd, ph, pw = patch.shape
        if pd != self.d or ph != self.h or pw != self.w:
            padded = np.zeros((self.d, self.h, self.w), dtype=np.float32)
            sd = (self.d - pd) // 2
            sh = (self.h - ph) // 2
            sw = (self.w - pw) // 2
            padded[sd:sd+pd, sh:sh+ph, sw:sw+pw] = patch
            patch = padded

        # 4) Tensor'a dönüştür ve kanal ekseni ekle: (1, D, H, W)
        img_tensor = torch.from_numpy(patch).unsqueeze(0)

        # 5) Augmentasyon (varsa)
        if self.transform and self.is_training:
            img_tensor = self.transform(img_tensor)

        label = torch.tensor(self.labels[idx], dtype=torch.long)
        return img_tensor, label

# ---------------------------------------
# 6) Gelişmiş 3D Augmentasyon
# ---------------------------------------
class AdvancedAugment3D:
    def __init__(self):
        self.spatial_transform = A.Compose([
            A.RandomRotate90(p=0.5),
            A.HorizontalFlip(p=0.5),
            A.VerticalFlip(p=0.5),
            A.Affine(shift_limit=0.1, scale_limit=0.1, rotate_limit=15, p=0.5),
            A.RandomBrightnessContrast(p=0.5),
            A.GaussNoise(p=0.3),
            A.GaussianBlur(p=0.3),
            ToTensorV2()
        ])

    def __call__(self, x):
        # x: torch tensor shape (1, D, H, W)
        if torch.rand(1) > 0.5:
            x = torch.flip(x, dims=[1])  # flip depth
        if torch.rand(1) > 0.5:
            x = torch.flip(x, dims=[2])  # flip height
        if torch.rand(1) > 0.5:
            x = torch.flip(x, dims=[3])  # flip width
        
        # Add random noise
        noise = torch.randn_like(x) * 0.02
        x = x + noise
        
        # Add random brightness/contrast
        brightness = 1.0 + (torch.rand(1) * 0.2 - 0.1)
        contrast = 1.0 + (torch.rand(1) * 0.2 - 0.1)
        x = (x - 0.5) * contrast + 0.5
        x = x * brightness
        
        x = torch.clamp(x, 0.0, 1.0)
        return x

# ---------------------------------------
# 7) Gelişmiş ResNet3D Mimarisi
# ---------------------------------------
class ResNet3DClassifier(nn.Module):
    def __init__(self, num_classes=3, pretrained=True):
        super().__init__()
        # Video ResNet18 (r3d_18) kullanıyoruz
        self.resnet3d = video_models.r3d_18(weights='DEFAULT' if pretrained else None)
        
        # İlk conv katmanını tek kanallı MR hacmi için güncelle
        self.resnet3d.stem[0] = nn.Conv3d(
            in_channels=1, out_channels=64, kernel_size=(3,7,7),
            stride=(1,2,2), padding=(1,3,3), bias=False
        )
        
        # Dropout ekle
        self.dropout = nn.Dropout(0.5)
        
        # Son fc katmanını yeniden tanımla
        self.resnet3d.fc = nn.Sequential(
            nn.Linear(self.resnet3d.fc.in_features, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.resnet3d(x)
        x = self.dropout(x)
        return x

# ---------------------------------------
# 8) Sabit Test Dataset & DataLoader
# ---------------------------------------
test_ds = ADNI_ResNet3D_Dataset(
    paths_test, labels_test,
    transform=None,
    target_depth=64, target_height=112, target_width=112,
    is_training=False
)
batch_size = 2  # Batch size'ı düşürdük
test_loader = DataLoader(
    test_ds, batch_size=batch_size,
    shuffle=False, num_workers=2,  # Worker sayısını düşürdük
    pin_memory=True, prefetch_factor=2
)

# ---------------------------------------
# 9) K-Fold Üzerinde Eğitim ve Değerlendirme
# ---------------------------------------
for fold, (train_index, val_index) in enumerate(
        skf.split(paths_trainval, labels_trainval), start=1
    ):

    print(f"\n========== Fold {fold}/{n_splits} ==========")

    # 9.1) Fold'a özel Train/Val Veri Ayırma
    paths_train = [paths_trainval[i] for i in train_index]
    labels_train = [labels_trainval[i] for i in train_index]
    paths_val   = [paths_trainval[i] for i in val_index]
    labels_val  = [labels_trainval[i] for i in val_index]
    print(f"Fold {fold} - Train sayısı: {len(paths_train)}, Val sayısı: {len(paths_val)}")

    # 9.2) Class Weights ve Weighted Sampler
    counter_train = Counter(labels_train)
    total_train   = sum(counter_train.values())
    class_weights = [total_train / counter_train[i] for i in range(3)]
    class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(DEVICE)
    print(f"Fold {fold} - Class weights:", class_weights_tensor.cpu().numpy())

    sample_weights = [class_weights[label] for label in labels_train]
    train_sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)

    # 9.3) Dataset ve DataLoader'lar
    train_ds = ADNI_ResNet3D_Dataset(
        paths_train, labels_train,
        transform=AdvancedAugment3D(),
        target_depth=64, target_height=112, target_width=112,
        is_training=True
    )
    val_ds = ADNI_ResNet3D_Dataset(
        paths_val, labels_val,
        transform=None,
        target_depth=64, target_height=112, target_width=112,
        is_training=False
    )

    train_loader = DataLoader(
        train_ds, batch_size=batch_size,
        sampler=train_sampler, num_workers=2,  # Worker sayısını düşürdük
        pin_memory=True, prefetch_factor=2
    )
    val_loader = DataLoader(
        val_ds, batch_size=batch_size,
        shuffle=False, num_workers=2,  # Worker sayısını düşürdük
        pin_memory=True, prefetch_factor=2
    )
    print(f"Fold {fold} - Train batch sayısı: {len(train_loader)}, Val batch sayısı: {len(val_loader)}")

    # 9.4) Model, Loss, Optimizer, Scheduler, AMP
    model = ResNet3DClassifier(num_classes=3, pretrained=True).to(DEVICE)
    
    # Model optimizasyonları
    torch.cuda.empty_cache()  # GPU belleğini temizle
    if hasattr(torch.cuda, 'amp'):
        scaler = GradScaler()
    else:
        scaler = None

    cls_loss_fn = CrossEntropyLoss(weight=class_weights_tensor)
    optimizer = torch.optim.AdamW(model.parameters(), lr=5e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=5e-4,
        epochs=20,
        steps_per_epoch=len(train_loader),
        pct_start=0.3,
        anneal_strategy='cos'
    )

    # 9.5) Eğitim Döngüsü
    num_epochs = 20  # 5'ten 20'ye çıkardık
    best_val_acc = 0.0
    patience = 5  # 3'ten 5'e çıkardık
    epochs_no_improve = 0
    early_stop = False

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []

    for epoch in range(1, num_epochs + 1):
        if early_stop:
            print(f"Epoch {epoch} itibarıyla Early Stopping devreye girdi. Eğitim durduruldu.")
            break

        # --------- TRAIN ---------
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0

        for imgs, labels in tqdm(train_loader, desc=f"Fold {fold} - Epoch {epoch}/{num_epochs} - Training"):
            imgs = imgs.to(DEVICE, non_blocking=True)
            labels = labels.to(DEVICE, non_blocking=True)

            optimizer.zero_grad(set_to_none=True)  # Daha verimli bellek kullanımı
            
            if scaler is not None:
                with autocast(device_type='cuda', dtype=torch.float16):
                    logits = model(imgs)
                    loss = cls_loss_fn(logits, labels)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            else:
                logits = model(imgs)
                loss = cls_loss_fn(logits, labels)
                loss.backward()
                optimizer.step()

            scheduler.step()

            running_loss += loss.item()
            preds = logits.argmax(dim=1)
            correct_train += (preds == labels).sum().item()
            total_train += labels.size(0)

            # Bellek temizliği
            del logits, loss, preds
            torch.cuda.empty_cache()

        avg_train_loss = running_loss / len(train_loader)
        train_acc = correct_train / total_train
        print(f"Fold {fold} - Epoch {epoch} - Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.4f}")

        train_losses.append(avg_train_loss)
        train_accs.append(train_acc)

        # --------- VALIDATION ---------
        model.eval()
        running_val_loss = 0.0
        correct_val = 0
        total_val = 0

        with torch.no_grad():
            for imgs, labels in tqdm(val_loader, desc=f"Fold {fold} - Epoch {epoch}/{num_epochs} - Validation"):
                imgs = imgs.to(DEVICE)
                labels = labels.to(DEVICE)

                if scaler is not None:
                    with autocast(device_type='cuda', dtype=torch.float16):
                        logits = model(imgs)
                        loss = cls_loss_fn(logits, labels)
                else:
                    logits = model(imgs)
                    loss = cls_loss_fn(logits, labels)

                running_val_loss += loss.item()
                preds_val = logits.argmax(dim=1)
                correct_val += (preds_val == labels).sum().item()
                total_val += labels.size(0)

                # Bellek temizliği
                del logits, loss, preds_val
                torch.cuda.empty_cache()

        avg_val_loss = running_val_loss / len(val_loader)
        val_acc = correct_val / total_val
        print(f"Fold {fold} - Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.4f}")

        val_losses.append(avg_val_loss)
        val_accs.append(val_acc)

        if val_acc > best_val_acc + 1e-4:
            best_val_acc = val_acc
            epochs_no_improve = 0
            fold_model_path = f'/kaggle/working/best_resnet3d_fold{fold}.pth'
            torch.save(model.state_dict(), fold_model_path)
            print(f"-- Fold {fold}: Yeni en iyi model kaydedildi (val_acc yükseldi) --")
        else:
            epochs_no_improve += 1
            print(f"Fold {fold} - Val acc iyileşmedi (epochs_no_improve: {epochs_no_improve}/{patience})")
            if epochs_no_improve >= patience:
                early_stop = True
                print(f"Fold {fold} - Early stopping patience ({patience}) aşıldı. Eğitim durduruluyor.")

        # Her epoch sonunda bellek temizliği
        torch.cuda.empty_cache()

    # --------- 9.6) Learning Curve Görselleştirme ---------
    plt.figure(figsize=(8, 4))
    plt.plot(range(1, len(train_losses) + 1), train_losses, label='Train Loss')
    plt.plot(range(1, len(val_losses) + 1), val_losses, label='Val Loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss')
    plt.legend(); plt.title(f'Fold {fold} - Loss Learning Curve')
    plt.show()

    plt.figure(figsize=(8, 4))
    plt.plot(range(1, len(train_accs) + 1), train_accs, label='Train Acc')
    plt.plot(range(1, len(val_accs) + 1), val_accs, label='Val Acc')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy')
    plt.legend(); plt.title(f'Fold {fold} - Accuracy Learning Curve')
    plt.show()

    # ---------------------------------------
    # 10) Test Aşaması
    # ---------------------------------------
    best_model_path = f'/kaggle/working/best_resnet3d_fold{fold}.pth'
    model.load_state_dict(torch.load(best_model_path, map_location=DEVICE))
    model.eval()

    all_targets = []
    all_preds   = []

    with torch.no_grad():
        for imgs, labels in tqdm(test_loader, desc=f"Fold {fold} - Testing"):
            imgs   = imgs.to(DEVICE)
            labels = labels.to(DEVICE)
            with autocast(device_type='cuda', dtype=torch.float16):
                logits = model(imgs)
            preds = logits.argmax(dim=1)
            all_targets.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())

    test_acc = np.mean(np.array(all_preds) == np.array(all_targets))
    print(f"\nFold {fold} - Test Accuracy: {test_acc:.4f}\n")
    print(f"Fold {fold} - Classification Report:")
    print(classification_report(all_targets, all_preds, target_names=["CN", "MCI", "AD"]))

    cm = confusion_matrix(all_targets, all_preds)
    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm, annot=True, fmt='d',
        xticklabels=["CN", "MCI", "AD"],
        yticklabels=["CN", "MCI", "AD"],
        cmap='Blues'
    )
    plt.xlabel("Predicted"); plt.ylabel("True"); plt.title(f"Fold {fold} - Confusion Matrix")
    plt.show()
```
